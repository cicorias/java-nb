```java
//question 2.

//Result O(n^2)

int myCounter = 0;

//first setup the method and kill any console output (noise...)
public static int f(int N) { 
    //System.out.println(N);
    myCounter++;
    // Hint: check how many println you will get
    if (N <= 1) { return 1; } 
    else { return f(N - 1); }
}

//create a loop to generate a data set to graph.

int[] xData = new int[100];
int[] yData = new int[100];

for(int x = 1 ; x < xData.length; x++) {
    
    int y = f(x);
    if (x % 10 == 0) System.out.println("x:= " + x + "  y:= " + y + "   counter:= " + myCounter);
    //now store variables.
    xData[x] = x;
    yData[x] = myCounter;
    //yData[x] = y / x;
    
}

chartData(xData, yData);


```

```java

myCounter = 0;
void example2(int n) {
    for (int i = 0; i < n; i++) // loop 1
    for (int j = i+1; j > i; j--) // loop 2
    for (int k = n; k > j; k--) // loop 3
        ++myCounter;
    //System.out.println("*");
}



int[] xData = new int[100];
int[] yData = new int[100];


for(int x = 1 ; x < xData.length; x++) {
    
    example2(x);
    if (x % 10 == 0) System.out.println("x:= " + x + "counter:= " + myCounter);
    //now store variables.
    xData[x] = x;
    yData[x] = myCounter;
    //yData[x] = y / x;
}

chartData(xData, yData);
```
