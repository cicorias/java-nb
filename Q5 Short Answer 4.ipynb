```java

/// Big O --- O(n^2)
int myCounter = 1;

void question4(int n) {
    int count = 0;
    int i = 1;
    while(i < n){
        for (int j = 1; j < n*n; j *= n) {
              count++;
              ++myCounter;
        }
        i *= 2;
        //++myCounter;
    }
    //System.out.println(count);
    // Hint: check the count you will get
}

int[] xData = new int[100];
int[] yData = new int[100];

for(int x = 1 ; x < xData.length; x++) {
    
    question4(x);
    if (x % 10 == 0) System.out.println("x:= " + x + "   counter:= " + myCounter);
    //now store variables.
    xData[x] = x;
    yData[x] = myCounter;
    //yData[x] = y / x;
    
}

chartData(xData, yData);

```
