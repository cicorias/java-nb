```java

/// Big O --- O(n^2)
int innerCounter = 1;
int outerCounter = 1;

int question4(int n) {
    int count = 0;
    int i = 1;
    while(i < n){
        for (int j = 1; j < n*n; j *= n) {
              //System.out.println("i: " + i + " j: " + j);
              count++;
              ++innerCounter;
        }
        i *= 2;
        ++outerCounter;
    }
    //System.out.println(count);
    // Hint: check the count you will get
    return count;
}


int points = 10000;

int[] xData = new int[points];
int[] yData = new int[points];
int[] y2Data = new int[points];
int[] y3Data = new int[points];

for(int x = 1 ; x < xData.length; x++) {
    innerCounter = 1;
    outerCounter = 1;
    int count = question4(x);
    if (x % (points / 10) == 0) System.out.println("x:= " + x + "   innercounter:= " + innerCounter + "  outercounter:= " + outerCounter + "   count: " + count);
    //if (x % 10 == 0) System.out.println("x:= " + x + "   outercounter:= " + outerCounter);
    //now store variables.
    xData[x] = x;
    yData[x] = outerCounter;
    y2Data[x] = innerCounter;
    y3Data[x] = outerCounter + innerCounter;
    //yData[x] = y / x;
    
}
```
